### Window Functions

```python
w = pd.DataFrame({'c':[1,2,3,4,5,6,7,7,88,67,8888],'d':[1,2,3,4,5,6,7,7,88,67,8888]})
```

```python
w
```

```python
w.rolling(window =1).mean() #mean of just current row
```

```python
w['c'].rolling(window =2).mean() #mean of current and previous row each row
```

```python
w.rolling(window =2).mean() #mean of current and previous row each row
```

```python
w.rolling(window =3).min() #mean of current and previous 2 rows each row
```

```python
w.cumsum() # does cummulative sum as we go down the row
```
