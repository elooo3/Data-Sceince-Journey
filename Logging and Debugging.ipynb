Logging is particularly useful in production especially to debug code whenever errors occur.
When writing production code we often don't write print statements as nobody can see compiler and 
they dissapear once script is run. Instead we can log useful bits of information to files which can later help us 
debug issues or trace history of scripts

Logging is one of those good code practices and nobody needs to tell you to log

Lets see how we can apply some logging to above

```python
l = [1,2,1+j,3,4,5,[True,1,'ope', 4, 1+2j], 'costa', 'opta']
lint = []
lstr = []
logging.basicConfig(filename = 'Testlogicfile.log', level = logging.DEBUG, format = '%(asctime)s %(message)s')

logging.info('Start of loop, we are now looping through {}'.format(l))
for i in l:
    logging.info('current value if l is {} '.format(i))
    if type(i) == list:
        logging.info('I have spotted a list type in l, the item is {}'.format(i))
        for j in i:
            logging.info('Now iterating through the list {}'.format(i))
            if type(j) == int:
                logging.info('currently in {r}, integer found is {t}'.format(r=i, t=j))
                lint.append(j)
            elif type(j) == str:
                logging.info('currently in {r}, string found is {t}'.format(r=i, t=j))
                lstr.append(j)
            else:
                logging.info('currently in {r}, item neither integer nor string is {t}'.format(r=i, t=j))
                pass
    elif type(i) == int:
        logging.info('I have spotted an integer type in l, the item is {}'.format(i))
        lint.append(i)
        
    elif type(i) == str:
        logging.info('I have spotted a string type in l, the item is {}'.format(i))
        lstr.append(i)
        
    else:
        logging.info('This item is neither an integer nor a string, the item is {}'.format(i))
        pass
logging.info('end of loop, list of integers found is {c} while list of strings found is {d}'.format(c= lint, d= lstr))
```
