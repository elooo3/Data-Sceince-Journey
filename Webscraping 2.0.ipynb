```python
import logging as lg
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen #useful for opening urls
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
url = 'https://www.jumia.com.ng/catalog/?q=bag' #url we want to scrape from
url2 = 'https://www.jumia.com.ng'
reviews = []
try:
    # Type search string and build complete url for scraping
    total_pages = 2+1
    reviewsPages = 2 + 1  # no of pages to scrap plus 1

    for i in range(1, total_pages):
      flipkart_url = url + "&page=" + str(i)
      flipkartPage =  requests.get(flipkart_url, headers=headers)  # reading the webpage
      flipkart_html = bs(flipkartPage.content,features="lxml")  # parsing the webpage as HTML
      # big boxes in each page
      big_boxes = flipkart_html.find_all("article", {"class": "prd _fb col c-prd"})
      for box in range(len(big_boxes)):
        productLink = url2 + big_boxes[box].find_all('a')[1]['href']  # extracting the actual product link
        #print(productLink)
        prodRes = requests.get(productLink, headers=headers)  # getting the product page from serve
        prod_html = bs(prodRes.content, features="lxml")  # parsing the product page as HTML
        plinks2 = prod_html.find_all("section", {"class": "card aim -mtm"})
        try:
            all_reviews_link = plinks2[0].find('a')['href']
            productAllReviewLink = url2 + all_reviews_link # full url to all comments for this product
            #print(productAllReviewLink)
        except:
            continue
        # Visit each of the review pages individually and scrape
        for k in range(1, reviewsPages):
            if k>1:
                review_urls = [productAllReviewLink + '&page=' + str(k)]
            else:
                review_urls = [productAllReviewLink]
        # create another for loop for each review page
           # print(review_urls)
            for review_url in review_urls:
                try:
                    revRes = requests.get(review_url, headers=headers)  # getting the review_url for an individual page
                    review_html = bs(revRes.content,features="lxml")  # parsing the review page as HTML
                    comment_boxes = review_html.find_all("article", {"class": "-pvs -hr _bet"})
                  #reviews = []  # initializing an empty list for reviews
                  #Iterate through the comment boxes to get the details of customer and their comments
                    for comments in comment_boxes: 
                        try:
                            name = comments.find_all('div')[2].div.find_all('span')[1].text 
                        except:
                            name = 'No Name'
                        try:
                            rating = comments.div.text
                        except:
                            rating = "No Rating"
                        try:
                            commentDate = comments.find_all('div')[2].div.span.text
                        except:
                            commentDate = "No Date"
                        try:
                            commentHead = comments.h3.text 
                        except:
                            commentHead = 'No Comment Heading'
                        try:
                            custComment = comments.p.text
                        except:
                            custComment = 'No Customer Comment'

                        my_dict = {"Product": 'Bag', "Name": name, "Rating": rating,
                                "Comment": custComment, "Date": commentDate, "CommentHead": commentHead}  # saving that detail to a dictionary
                        #x = table.insert_one(my_dict)  # insertig the dictionary containing the rview comments to the collection
                        reviews.append(my_dict)  # appending the comments to the review list
                        print(len(reviews))
                except:
                    continue
except: print("something is wrong")
```
