```python
s = [['name','email','tel'],['ope','ope@mail.com','12345'],['josh','josh@mail.com','67893']]
```

```python
import csv
```

```python
# store s as a csv
with open('test.csv','w') as f:
    x = csv.writer(f)
    for i in s:
        x.writerow(i)
```

```python
#another way to read from files
with open('test.csv','r') as t:
    print(t.read())
```

```python
t.close()
```

```python
#another way to read from files
with open('test.csv','r') as t:
    r = csv.reader(t)
    for i in r:
        print(i)
```

```python
import io
```

```python
#Buffered Read and Write mainlt for larger files
#another way to write from files
with open('test1.txt','wb') as t:
    r = io.BufferedWriter(t) #buffer write for performing write to file in buffer chunks
    r.write(b'Many of you will be familiar with running python code in Jupyter notebook and using markdown to create descriptions and text. Some may not know that you can run shell commands from the notebook itself. \n') #writing to file
    r.write(b'this is a new line')
    r.flush() #close and commit changes
```

```python
#Buffered Read and Write mainly for larger files
#another way to read from files
with open('test1.txt','rb') as t:
    r = io.BufferedReader(t) #buffer read for reading from file in buffer chunks
    f = r.read(30) #you specify the amount of bytes you want to read from file within the read function
    print(f)
```
