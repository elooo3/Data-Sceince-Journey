```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a dataset comprising of 50 data points
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y) # plotting points as a scatter plot
plt.xlabel('x axis') # labeling x-axis with text within brackets
plt.ylabel('y axis') # labeling y-axis with text within brackets
plt.title('plotting x vs y') # setting uour title
```

```python
plt.scatter(x,y, c= '#a1ff1d') # changing colour of points using hex colour code
```

```python
plt.scatter(x,y, c= '#a1ff1d', alpha = 0.5) # changing intensity of colours using alpha
```

```python
plt.scatter(x,y, c= '#a1ff1d', alpha = 0.5, marker = 'p') # changing shape of markers
```

```python
# adding a grid
plt.scatter(x,y, c= '#a1ff1d', alpha = 0.5, marker = 'p') 
plt.grid()
```

```python
# create a dataset comprising of 50 data points
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
plt.scatter(x,y,z) # not a good way to visualize 3d data
```

```python
# for plotting a 3d graph
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.xlabel('x-axis') # label x-axis
plt.ylabel('y-axis') # label y-axis
plt.zlabel('z-axis') # there is no z label function
plt.title('3d graph') # adding a title
#plt.show()
```
