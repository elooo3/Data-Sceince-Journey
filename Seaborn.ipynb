```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x= iris.sepal_length, y = iris.petal_length)
```

```python
sns.distplot(iris['sepal_width'])
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x= tips['total_bill'], y = tips['tip'])
```

```python
tips.head() # show first 5 rows
```

```python
tips['smoker'].value_counts() # show the count of distinct values
```

```python
tips['size'].value_counts() # show the count of distinct values
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips,hue= 'smoker') # groups by smoker column
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips,hue= 'size') # groups by size column
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips,col= 'size') # groups by size column
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips,col= 'time') # groups by size column
```

```python
sns.relplot(x= 'total_bill', y = 'tip', data = tips,style = 'size' ,col= 'time') # we can see time and group size
```

```python
tips['time'].value_counts() # show the count of distinct values
```

```python
tips
```

```python
# cat plot is useful for plotting categorical data against numerical data
sns.catplot(x = 'day', y = 'tip', data = tips) # we can see how tips varied each day
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips) # we can see how tips aried each day
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips) # we can see how tips varied each day
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips) # joint plot combines a scatter plot and histogram
```

```python
sns.pairplot(tips)  #shows the relationship between all numeric data only on a plot
```

```python
sns.pairplot(iris)  #shows the relationship between all numeric data on a plot
```
